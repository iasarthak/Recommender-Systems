```python
pivot_table=dataset.pivot('user_id','article_id','rating')
pivot_table
```

Q1.

```python
total_ratings=pivot_table.count(axis=0)
total_ratings[total_ratings.values>0.05*907].sort_values(ascending=False).head(10)
```

Q2.

```python
mean_ratings=pivot_table.mean(axis=0)
mean_ratings[mean_ratings.values>1.5].sort_values(ascending=False).head(10)
```

Q3.

```python
weighted=(total_ratings*mean_ratings+mean_ratings.mean()*2)/(total_ratings+2)
weighted.sort_values(ascending=False).head(10)
```
